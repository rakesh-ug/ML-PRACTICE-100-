```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['cgpa'], df['package'], c='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
model.predict(X_test.iloc[0].values.reshape(1,1))   
```

```python
model.predict(X_test.iloc[2].values.reshape(1,1)) 
```

```python
plt.scatter(df['cgpa'], df['package'], c='red')
plt.plot(X_train, model.predict(X_train),c='blue')
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
m = model.coef_
```

```python
b = model.intercept_
```

```python
# y = mx + b

y = m*X_test.iloc[0,0] + b 
# or this way
y2 = m * X_test.values[[0]] + b
```
